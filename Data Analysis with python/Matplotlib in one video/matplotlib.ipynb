# 1. Importing matplotlib and some necessary libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
# for static plots
%matplotlib inline  
# %matplotlib  for intractive plots 
```

### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,label="sin")
plt.title("Sine Wave")
plt.plot(x,cos_y,label="cos")

plt.legend(loc = "upper right")
plt.title("Cosine Wave")
plt.show()
```

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')

plt.title("Cosine Wave")

plt.show()
```

# 2. Setting Styles to the Graph

### 2.1 One way to use styles is use plt.style.use('style_name'). List of available styles is:

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-dark')

plt.plot(x,sin_y,label = 'sin')
plt.plot(x,cos_y,label = 'cos')
plt.legend(loc = "lower right")
plt.title("Cosine & Sine Wave")

plt.show()
```

### 2.2 We can also increase size of the plot

```python
plt.figure(figsize=(18,5))

plt.style.use('grayscale')

plt.plot(x,sin_y,label = 'sin')
plt.plot(x,cos_y,label = 'cos')
plt.legend(loc = "lower right")
plt.title("Cosine & Sine Wave")

plt.show()
```

# 3. Saving the Plots in our Local Machine

```python
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

fig = plt.figure()

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

fig.savefig('Sine & Cosine.png')
```

# 4. Different Types of Plots

### 4.1 Line Plot

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x, y)


plt.title('Sine & Cosine Wave')

plt.show()
```

### 4.1.2 Multiple Line Plots

```python
x = np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,6))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin,'--',color = 'red',linewidth= 5,label = 'Sine ')
plt.plot(x, y_cos,'*',color = 'blue',linewidth= 3,label = 'cos ')

plt.legend(loc='lower right')
plt.title('Sine & Cosine Wave')

plt.show()
```

### 4.1.3 For real world datasets

```python
from sklearn.datasets import load_boston

dataset = load_boston()
target = dataset['target']
```

```python
plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')

plt.plot(target[:200], color = 'blue', linewidth = 1)

plt.show()
```

### 4.1.4 We can also use linestyle command. It has 4 main styles solid, dashed, dashdot and dotted

```python
x = np.linspace(0, 20, 21, dtype = 'int')

plt.figure(figsize = (15,8))

plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed') 
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')

plt.legend(loc = 'upper left')

plt.show()
```

### 4.1.5 Adjusting the Plot Axes

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,20)
plt.ylim(-2.5,2.5)

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

### Can also be done using plt.axis([xlim,ylim])

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim([0,10])
plt.ylim([-2.5,2.5])

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

### 4.1.6 Using .set to combine all of these commands

```python
x = np.linspace(0, 4*np.pi, 100)

fig = plt.axes()

fig.plot(x, np.sin(x))

fig.set(xlim = (0,10), ylim = (-2,2), xlabel = 'x', ylabel = 'sin(x)', title = 'Sie Wave')

fig.plot()
```

### 4.2.3 On Real World Dataset

```python
from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

plt.scatter(x[:,0], x[:,1],c='blue',s=50,alpha=0.9)

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])
```

```python
plt.scatter(x[:,0], x[:,1],c='blue',s=50,alpha=0.9,marker = 'x')

plt.xlabel(iris['feature_names'][0])
plt.ylabel(iris['feature_names'][1])

```

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes  = rng.rand(100) * 1000

plt.scatter(x,y, c = colors, s = sizes , alpha = .7, cmap = 'viridis', edgecolors = 'black')

plt.title('Random Scatter Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.colorbar()

plt.show()
```
