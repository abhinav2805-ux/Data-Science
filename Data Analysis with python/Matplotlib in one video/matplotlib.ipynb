# 1. Importing matplotlib and some necessary libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
# for static plots
%matplotlib inline  
# %matplotlib  for intractive plots 
```

### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,label="sin")
plt.title("Sine Wave")
plt.plot(x,cos_y,label="cos")

plt.legend(loc = "upper right")
plt.title("Cosine Wave")
plt.show()
```

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')

plt.title("Cosine Wave")

plt.show()
```

# 2. Setting Styles to the Graph

### 2.1 One way to use styles is use plt.style.use('style_name'). List of available styles is:

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-dark')

plt.plot(x,sin_y,label = 'sin')
plt.plot(x,cos_y,label = 'cos')
plt.legend(loc = "lower right")
plt.title("Cosine & Sine Wave")

plt.show()
```

### 2.2 We can also increase size of the plot

```python
plt.figure(figsize=(18,5))

plt.style.use('grayscale')

plt.plot(x,sin_y,label = 'sin')
plt.plot(x,cos_y,label = 'cos')
plt.legend(loc = "lower right")
plt.title("Cosine & Sine Wave")

plt.show()
```

# 3. Saving the Plots in our Local Machine

```python
plt.figure(figsize = (18,5))

plt.style.use('grayscale')

fig = plt.figure()

plt.plot(x, sin_y,label = 'Sine')
plt.plot(x, cos_y,label = 'Cosine')

plt.title('Sine & Cosine Wave')

fig.savefig('Sine & Cosine.png')
```
