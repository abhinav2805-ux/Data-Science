# 1. Importing matplotlib and some necessary libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
# for static plots
%matplotlib inline  
# %matplotlib  for intractive plots 
```

### 1.1 Generating Sin and Cos Graph using Matplotlib

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,label="sin")
plt.title("Sine Wave")
plt.plot(x,cos_y,label="cos")

plt.legend(loc = "upper right")
plt.title("Cosine Wave")
plt.show()
```

```python
x = np.linspace(0,2*np.pi,50)

sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x,sin_y,'*')
plt.plot(x,cos_y,'--')

plt.title("Cosine Wave")

plt.show()
```

# 4. Different Types of Plots

### 4.1 Line Plot

```python
x = np.linspace(0,4*np.pi,100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x, y)


plt.title('Sine & Cosine Wave')

plt.show()
```

### 4.1.2 Multiple Line Plots

```python
x = np.linspace(0,4*np.pi,100)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(15,6))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin,'--',color = 'red',linewidth= 5,label = 'Sine ')
plt.plot(x, y_cos,'*',color = 'blue',linewidth= 3,label = 'cos ')

plt.legend(loc='lower right')
plt.title('Sine & Cosine Wave')

plt.show()
```

### 4.1.3 For real world datasets

```python
from sklearn.datasets import load_boston

dataset = load_boston()
target = dataset['target']
```

```python
plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')

plt.plot(target[:200], color = 'blue', linewidth = 1)

plt.show()
```

### 4.1.4 We can also use linestyle command. It has 4 main styles solid, dashed, dashdot and dotted

```python
x = np.linspace(0, 20, 21, dtype = 'int')

plt.figure(figsize = (15,8))

plt.plot(x, x + 0, linestyle = 'solid'  , label = 'Solid')
plt.plot(x, x + 1, linestyle = 'dashed' , label = 'Dashed') 
plt.plot(x, x + 2, linestyle = 'dashdot', label = 'DashDot')
plt.plot(x, x + 3, linestyle = 'dotted' , label = 'Dotted')

plt.legend(loc = 'upper left')

plt.show()
```

### 4.1.5 Adjusting the Plot Axes

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim(0,20)
plt.ylim(-2.5,2.5)

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

### Can also be done using plt.axis([xlim,ylim])

```python
x = np.linspace(0, 4*np.pi, 100)

plt.xlim([0,10])
plt.ylim([-2.5,2.5])

sin_y = np.sin(x)
plt.plot(x, sin_y)
plt.title('Sine Wave')

plt.show()
```

### 4.1.6 Using .set to combine all of these commands

```python
x = np.linspace(0, 4*np.pi, 100)

fig = plt.axes()

fig.plot(x, np.sin(x))

fig.set(xlim = (0,10), ylim = (-2,2), xlabel = 'x', ylabel = 'sin(x)', title = 'Sie Wave')

fig.plot()
```

### 4.6 Multiple Subplots

##### index based

```python
x = np.linspace(0,10,50)

fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x,y)
ax[0,0].set_title('Sin Wave')


y = np.cos(x)
ax[0,1].plot(x,y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x,y)
ax[1,0].set_title('Tan Wave')


y = np.random.randn(1000)
ax[1,1].hist(y,bins = 40)
ax[1,1].set_title('Random Distribution')


plt.tight_layout()
plt.show()
```

##### loop based

```python
plt.subplots_adjust(hspace = .4, wspace = .4)

for i in range(1,7):
    
    plt.subplot(2,3,i)
    plt.text(0.5 , 0.5, str((2,3,i)), fontsize = 18, ha = 'center')
```

```python
x = np.linspace(0,10,50)
y = np.sin(x)


fig , (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (18,6))
ax1.plot(x,y)
ax2.plot(x,y**2)
ax3.plot(x,y**3)

ax1.set_title('Sin Wave')
ax2.set_title('Sin Wave - Square')
ax3.set_title('Sin Wave - Cube')

plt.tight_layout()

plt.show()
```
